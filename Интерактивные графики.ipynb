```python
import pandas as pd
import plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
from sklearn.preprocessing import OneHotEncoder
import plotly.figure_factory as ff
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
from datetime import datetime
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
app = dash.Dash(
  __name__
)
```

```python
df = pd.read_excel('ВкусВилл,Теребиленко Матвей,тестовое задание.xlsx',sheet_name='Данные')
```

```python
df = df.drop(['Date(молния)','Date(высокий)','Date(средний)','Date(низкий)','Data for graphics','product_team',
              'dev_team_lead','stage_deadline','sprint_name','sprint_date_start','sprint_date_finish',
             'parent_id','dev_team'],axis=1)
```

```python
df=df.dropna()
```

```python
app = dash.Dash(
  __name__
)


app.layout = html.Div([
                html.Div([
                    html.H1('Интерактивные графики для анализа данных ВкусВилл'),
                    
                    html.P('Анализ данных о багах.'
                        'Используйте филтры,чтобы увидеть результат.')],
                    style = {'borderBottom': 'thin lightgrey solid','backgroundColor': 'rgb(250, 250, 250)',
                             'padding': '10px 5px' }) ,
html.Div([
    html.Div([
        html.Label('Ответственные команды'),
        dcc.Dropdown(
            id = 'crossfilter-executor_team',
            options = [{'label': i, 'value': i} for i in df['executor_team'].unique()],
            value = ['Группа ВкусВилл Интегратор. TEAM 3'],
            multi = True
        )
    ],
    style = {'width': '49%', 'display': 'inline-block'})
]),

html.Div([
    html.Label("Статус задачи"),
    dcc.Dropdown(
        id = 'crossfilter-task_status',
        options = [{'label': i, 'value': i} for i in df['task_status'].unique()],
        value = ['В работе','Отменено',"Закрыто"],
        multi = True
        )
]),
html.Div([
    html.Label("Приоритетность задачи"),
    dcc.Dropdown(
        id = 'crossfilter-priority_general',
        options = [{'label': i, 'value': i} for i in df['priority_general'].unique()],
        value = ['3. Средний','4. Низкий'],
        multi = True
    )
]),
html.Div(
    dcc.Graph(id = 'histogram'),
    style = {'width': '101%', 'display': 'inline-block'}
),
html.Div(
    dcc.Graph(id = 'scatter-plot'),
    style = {'width': '101%', 'display': 'inline-block'}
),

    html.Div(
    dcc.Graph(id = 'histogram-2'),
    style = {'width': '101%', 'display': 'inline-block','stackgroup':'one'}
    
),
    html.Div(
    dcc.Graph(id = 'ecdf'),
    style = {'width': '101%', 'display': 'inline-block','stackgroup':'one'}
)
])    

@app.callback(
    Output('histogram','figure'),
    [Input('crossfilter-executor_team','value'),
    Input('crossfilter-task_status','value'),
    Input('crossfilter-priority_general','value')])
def update_stacked_area(executor_team,task_status,priority_general):
    
    filtered_data = df[(df['executor_team'].isin(executor_team))&(df['task_status'].isin(task_status))&
                      (df['priority_general'].isin(priority_general))]
    figure_hist = px.histogram(filtered_data,
        x = "task_status", color = "priority_general",
        title = "Статус задач в соотношении с приоритетностью")
    figure_hist.update_layout(yaxis_title = "Кол-во")
    return figure_hist
    

@app.callback(
    Output('scatter-plot','figure'),
    [Input('crossfilter-executor_team','value'),
    Input('crossfilter-task_status','value'),
    Input('crossfilter-priority_general','value')])
def update_stacked_area(executor_team,task_status,priority_general):
    
    filtered_data = df[(df['executor_team'].isin(executor_team))&(df['task_status'].isin(task_status))&
                      (df['priority_general'].isin(priority_general))]    
    
    figure = px.scatter(filtered_data,
        x = "task_status", color = "priority_general",
        title = "Статус задач в соотношении с приоритетностью")
    figure.update_layout(yaxis_title = "Кол-во")
    return figure
    
@app.callback(
    Output('histogram-2','figure'),
    [Input('crossfilter-executor_team','value'),
    Input('crossfilter-task_status','value'),
    Input('crossfilter-priority_general','value')])
def update_stacked_area(executor_team,task_status,priority_general):
    
    filtered_data = df[(df['executor_team'].isin(executor_team))&(df['task_status'].isin(task_status))&
                      (df['priority_general'].isin(priority_general))]    
    
    figure = px.histogram(filtered_data,
        x = "executor_team", color = "priority_general",
        title = "Как команды справляются с очередью")
    figure.update_layout(yaxis_title = "Кол-во")
    return figure
    
@app.callback(
    Output('ecdf','figure'),
    [Input('crossfilter-executor_team','value'),
    Input('crossfilter-task_status','value'),
    Input('crossfilter-priority_general','value')])
def update_stacked_area(executor_team,task_status,priority_general):
    
    filtered_data = df[(df['executor_team'].isin(executor_team))&(df['task_status'].isin(task_status))&
                      (df['priority_general'].isin(priority_general))&(df['DateDiff'])]    
    
    figure = px.ecdf(filtered_data,
        x = "DateDiff", color = "executor_team",
        title = "За сколько команды выполняют определенные типы задачи")
    return figure



if __name__ == '__main__':
    app.run_server(debug=True, port=1213)
```
